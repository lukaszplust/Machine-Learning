Wytnij wszystkie wiersze obiektu df dla których kolumna weight nie przyjmuje wartość np.nan. Na tak otrzymanym obiekcie policz wartość średnią dla kolumn numerycznych price oraz weight. Wynik wydrukuj do konsoli.

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer


data = {
    'size': ['XL', 'L', 'M', np.nan, 'M', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red', 'green'],
    'gender': ['female', 'male', np.nan, 'female', 'female', 'male'],
    'price': [199.0, 89.0, np.nan, 129.0, 79.0, 89.0],
    'weight': [500, 450, 300, np.nan, 410, np.nan],
    'bought': ['yes', 'no', 'yes', 'no', 'yes', 'no']
}

df = pd.DataFrame(data=data)
```

```python
df
```

```python
df[~df['weight'].isnull()]
```

```python
#~df['weight'].isnull() - tworzy maskę logiczną, która wybiera wiersze,
#w których kolumna "weight" nie ma wartości null (NaN).

#średnia wartość liczbową w kolumnach, które zawierają wartości liczbowe w wybranych wierszach -> mean(numeric_only = True)
```

```python
print(df[~df['weight'].isnull()].mean(numeric_only=True))
```
