# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
#all rows without the last --> [:, :-1]
#all rows with only first column --> [:, :1]
#all rows with only last column --> [:, -1]
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
X[:5]
```

```python
y[:5]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Multiple Linear Regression

## Importing the libraries

## Importing the dataset

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
# X will be all variables without profit
X[:5]
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

# [3] oznacza, e do dekodowania uyjemy 3 kolumny (liczac od zera) czyli nazwy miast

#remainder='passthrough' oznacza, e wszystkie pozostae kolumny,
#ktre nie s poddane transformacji, bd przekazywane (passthrough) bez zmiany do wynikowej ramki danych
```

```python
X[:5]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Multiple Linear Regression model on the Training set

```python
X_train[:3]
```

```python
y_train[:3]
```

```python
X_test[:3]
```

```python
y_test[:3]
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Support Vector Regression (SVR)

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
len(y)
```

```python
y = y.reshape(len(y),1)
y
```

image.png

image.png

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
X
```

```python
y
```

## Training the SVR model on the whole dataset

image.png

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
```

## Predicting a new result

```python
sc_y.inverse_transform(regressor.predict(sc_X.transform([[6.5]])).reshape(-1,1))
```

## Visualising the SVR results

```python
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'red')
plt.plot(sc_X.inverse_transform(X), sc_y.inverse_transform(regressor.predict(X).reshape(-1,1)), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the SVR results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'red')
plt.plot(X_grid, sc_y.inverse_transform(regressor.predict(sc_X.transform(X_grid)).reshape(-1,1)), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Random Forest Regression

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Random Forest Regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 20, random_state = 0)
#n_estimators = model skada si z 10 drzew decyzyjnych
regressor.fit(X, y)
```

## Predicting a new result

```python
regressor.predict([[6.5]])
```

## Predictions

## Dependence on n_estimators value

```python
# Zakres wartoci n_estimators do przetestowania
n_estimators_values = [1, 5, 10, 50,200,400]

# Ustawienie wielkoci subplotw
plt.figure(figsize=(15, 10))

for i, n_estimators in enumerate(n_estimators_values, 1):
    # Inicjalizacja lasu losowego
    regressor = RandomForestRegressor(n_estimators=n_estimators, random_state=0)

    # Trenowanie modelu
    regressor.fit(X, y)

    # Predykcja na danych treningowych
    y_train_pred = regressor.predict(X)

    # Tworzenie osobnego subplotu
    plt.subplot(2, 3, i)

    # Wizualizacja rzeczywistych danych treningowych
    plt.scatter(X, y, color='blue', label='Actual', alpha=0.5)

    # Wizualizacja krzywej uczenia
    plt.plot(X, y_train_pred, label=f'n_estimators={n_estimators}', linewidth=2, color='red')

    plt.title(f'n_estimators={n_estimators}')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.legend()

# Dostosowanie uoenia subplotw
plt.tight_layout()
plt.show()
```

## Visualising the Random Forest Regression results (higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
